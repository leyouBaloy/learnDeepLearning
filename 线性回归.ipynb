```python
import numpy as np
import torch
from torch.utils import data
from d2l import torch as d2l
```

# 构造数据


```python
true_w = torch.tensor([2,-3.4])
true_b = 4.2
features,labels = d2l.synthetic_data(true_w,true_b,1000) #synthetic_data人工合成数据
features,labels
```

# 调用框架中现有的api来读取数据

```python
def load_array(data_arrays,batch_size,is_train=True):
    """构造一个pytorch数据迭代器。"""
    dataset = data.TensorDataset(*data_arrays)
    return data.DataLoader(dataset,batch_size,shuffle=is_train)

batch_size = 10
data_iter = load_array((features,labels),batch_size)

next(iter(data_iter))
```

# 使用框架的预定义好的层

```python
from torch import nn

net = nn.Sequential(nn.Linear(2,1)) #Sequential顺序模型，相当于把层按顺序放进一个list里面，（2,1）表示输入维度2，输出维度1
```

# 初始化模型参数

```python
net[0].weight.data.normal_(0,0.01) #net[0]是linear层，weight是权重w，data是真实数据，normal是正态分布，_是修改原数据
#均值为0，方差0.01
net[0].bias.data.fill_(0) #bias是偏差，设置为0
```

# 计算均方误差使用的是MSELoss类，也称平方范数

```python
loss = nn.MSELoss()
```

# 实例化SGD实例（SGD随机梯度下降

```python
trainer = torch.optim.SGD(net.parameters(), lr=0.01) #net.parameters是net中的所有参数，
#即w和b
```

# 训练过程

```python
num_epochs = 10 #迭代三个周期
for epoch in range(num_epochs):
    for X, y in data_iter:
        l = loss(net(X), y)
        trainer.zero_grad()
        l.backward()
        trainer.step()
    l = loss(net(features),labels)
    print(f"epoch {epoch+1}, loss {l:f}")
```
